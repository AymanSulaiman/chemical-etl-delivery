```python
import os
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("chemical_loading") \
    .config("spark.some.config.option", "some-value") \
    .config("spark.jars", "/Users/aymansulaiman/projects/personal/chemical-etl-delivery/jars/postgresql-42.2.6.jar")\
    .getOrCreate()
```

```python
df = spark.read.option("header",True).csv("../data/all_data_csv")
# df.show()
```

```python
df.write \
.format("jdbc") \
.option("url", "jdbc:postgresql:postgres") \
.option("dbtable", "chemical_table") \
.option("user", "") \
.option("password", "") \
.option("driver", "org.postgresql.Driver") \
.save()
```
